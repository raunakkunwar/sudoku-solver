```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Function to analyze
def func(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

# Function to measure the time taken
def measure_time(n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

# Range of input sizes
input_sizes = np.arange(1, 1001, 100)  # Small to large inputs
times = []

# Measure time for each input size
for n in input_sizes:
    times.append(measure_time(n))

# Fit the empirical data to a polynomial (quadratic)
poly_coeff = np.polyfit(input_sizes, times, 2)  # Fit to a quadratic polynomial
fitted_poly = np.poly1d(poly_coeff)

# Upper bound O(n^2)
upper_bound = lambda n: 1.1 * (poly_coeff[0] * n**2 + poly_coeff[1] * n + poly_coeff[2])

# Lower bound Omega(n^2)
lower_bound = lambda n: 0.9 * (poly_coeff[0] * n**2)

# Generate a finer range of input sizes for smooth plotting of the polynomial
fine_input_sizes = np.linspace(1, 1000, 500)

# Plotting the results
plt.figure(figsize=(12, 6))
#plt.plot(input_sizes, times, 'b-', label='Measured Time', linewidth=2)  # Solid line for measured time
plt.plot(input_sizes, times, marker='o', label='Measured Time')

plt.plot(fine_input_sizes, fitted_poly(fine_input_sizes), label='Fitted Polynomial (Theta)', color='green')
plt.plot(fine_input_sizes, upper_bound(fine_input_sizes), label='Upper Bound (O(n^2))', linestyle='--', color='red')
plt.plot(fine_input_sizes, lower_bound(fine_input_sizes), label='Lower Bound (Omega(n^2))', linestyle='--', color='blue')

# Labels and title
plt.xlabel('n (Input Size)')
plt.ylabel('Tn (Time Taken)')
plt.title('Tn vs n with Polynomial Fitting')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Function to analyze
def func(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

# Function to measure the time taken
def measure_time(n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

# Range of input sizes
input_sizes = np.arange(1, 1001, 100)  # Small to large inputs
times = []

# Measure time for each input size
for n in input_sizes:
    times.append(measure_time(n))

# Fit the empirical data to a polynomial (quadratic)
poly_coeff = np.polyfit(input_sizes, times, 2)  # Fit to a quadratic polynomial
fitted_poly = np.poly1d(poly_coeff)

# Upper bound O(n^2)
upper_bound = lambda n: 1.1 * (poly_coeff[0] * n**2 + poly_coeff[1] * n + poly_coeff[2])

# Lower bound Omega(n^2)
lower_bound = lambda n: 0.9 * (poly_coeff[0] * n**2)

# Generate a finer range of input sizes for smooth plotting of the polynomial
fine_input_sizes = np.linspace(1, 1000, 500)

# Specify the location of n_0
n_0 = 500  # Example value for n_0

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(input_sizes, times, marker='o', label='Measured Time')
plt.plot(fine_input_sizes, fitted_poly(fine_input_sizes), label='Fitted Polynomial (Theta)', color='green')
plt.plot(fine_input_sizes, upper_bound(fine_input_sizes), label='Upper Bound (O(n^2))', linestyle='--', color='red')
plt.plot(fine_input_sizes, lower_bound(fine_input_sizes), label='Lower Bound (Omega(n^2))', linestyle='--', color='blue')

# Add vertical dotted line at n_0
plt.axvline(x=n_0, color='purple', linestyle='--', label=f'n_0 = {n_0} (Threshold)')

# Labels and title
plt.xlabel('n (Input Size)')
plt.ylabel('Tn (Time Taken)')
plt.title('Tn vs n with Polynomial Fitting and Vertical Line at n_0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Function to analyze
def func(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

# Modified function to analyze
def func_modify(n):
    x = 1
    y = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
            y = i + j
    return x

# Function to measure the time taken for both functions
def measure_time(n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    time_func = end_time - start_time
    
    start_time = time.time()
    func_modify(n)
    end_time = time.time()
    time_func_modify = end_time - start_time
    
    return time_func, time_func_modify

# Range of input sizes
input_sizes = np.arange(1, 1001, 100)  # Small to large inputs
times_func = []
times_func_modify = []

# Measure time for each input size for both functions
for n in input_sizes:
    time_func, time_func_modify = measure_time(n)
    times_func.append(time_func)
    times_func_modify.append(time_func_modify)

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(input_sizes, times_func, marker='o', label='func Time', color='blue')
plt.plot(input_sizes, times_func_modify, marker='s', label='func_modify Time', color='orange')

# Labels and title
plt.xlabel('n (Input Size)')
plt.ylabel('Tn (Time Taken)')
plt.title('Tn vs n for func and func_modify')
plt.legend()
plt.grid(True)
plt.show()
```
